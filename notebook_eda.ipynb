```python
import string
#define a function to preprocess narration column
# Preprocess the 'narration' column
def preprocess_narration(narration):
    # Convert to lowercase
    narration = narration.lower()
    # Remove punctuation
    narration = narration.translate(str.maketrans('', '', string.punctuation))
    return narration
```

```python
p1_transaction['type'].value_counts()
```

```python
#applying the function
p1_transaction['narration'] = p1_transaction['narration'].apply(preprocess_narration)

# Define keywords for each category
categories = {
    'ecommerce': ['amazon', 'flipkart', 'ebay', 'shopping'],
    'fuel': ['fuel', 'petrol', 'gas', 'diesel','filling'],
    'food': ['restaurant', 'dining', 'cafe', 'food', 'groceries'],
    'atm': ['atm', 'cash', 'withdrawal'],
    'upi': ['upi', 'paytm', 'google pay', 'phonepe'],
    'salary':['salary']
}

# Function to classify transactions based on narration
def classify_transaction(narration):
    for category, keywords in categories.items():
        if any(keyword in narration for keyword in keywords):
            return category
    return 'other'  # Default category if no keywords match

p1_transaction['Category'] = p1_transaction['narration'].apply(classify_transaction)

# Analyze the expenses by category
expense_summary = p1_transaction[p1_transaction['type']=='DEBIT'].groupby('Category')['amount'].sum().reset_index()
# Analyze the income by category
income_summary = p1_transaction[p1_transaction['type']=='CREDIT'].groupby('Category')['amount'].sum().reset_index()

# Display the expense summary
print(expense_summary)
# Display the income summary
print(income_summary)
```

```python
# Use IQR method to identify outliers
Q1 = p1_transaction['amount'].quantile(0.25)
Q3 = p1_transaction['amount'].quantile(0.75)
IQR = Q3 - Q1
# Identify outliers
p1_transaction['is_suspicious'] = (p1_transaction['amount'] > (Q3 + 1.5 * IQR)) | (p1_transaction['amount'] < (Q1 - 1.5 * IQR))
# Display outliers
suspicious_transaction = p1_transaction[p1_transaction['is_suspicious']]
suspicious_transaction[suspicious_transaction['Transaction_Size']=='Large'].T[[291]]
```

```python
#Alert Generation
# Set thresholds
low_balance_threshold = 1000  
high_expenditure_threshold = 5000

#df of debit only
p1_ex=p1_transaction[p1_transaction['type']=='DEBIT']

# Generate low balance alerts
p1_transaction['low_balance_alert'] = p1_transaction['currentBalance'] < low_balance_threshold

# Generate high expenditure alerts
p1_ex['high_expenditure_alerts'] = p1_ex[p1_ex['type']=='DEBIT']['amount'] > high_expenditure_threshold

# Display alerts-1
low_balance_alerts = p1_transaction[p1_transaction['low_balance_alert']]
print("Low Balance Alerts:")
print(low_balance_alerts[['valueDate', 'currentBalance']].reset_index())

# Display alerts-2
High_Expenditure_alerts = p1_ex[p1_ex['high_expenditure_alerts']]
print("High Expenditure Alerts:")
print(High_Expenditure_alerts[['valueDate', 'amount']].reset_index())
```

```python
p3_new=p3.drop(['Surname','Age Group'],axis=1)
p3_new.head()
```

```python
cat_feats=['Geography','Gender']
final_p3=pd.get_dummies(p3_new,columns=cat_feats,drop_first=True)
final_p3.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=final_p3.drop('churned',axis=1)
y=final_p3['churned']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred=dtree.predict(X_test)
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
pred1=rfc.predict(X_test)
```

```python
print(classification_report(y_test,pred1))
print(confusion_matrix(y_test,pred1))
```
